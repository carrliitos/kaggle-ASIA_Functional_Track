## Investigation Kaggle Comment

> Hi! I noticed that many time-varying covariates in both the training dataset and the test dataset have value 9, which is not explained by the data dictionary, and beyond the grading scale of these covariates. I am wondering whether "9" denoted "data not able to assess".

Link: https://www.kaggle.com/competitions/asia-data-science-challenge-2025/discussion/570653

```python
train_features_df.head()
```

```python
features_only = train_features_df.drop(columns=["PID"])
```

```python
features_with_9_values = features_only.columns[(features_only == 9).any()].tolist()
```

```python
value_9_counts = features_only[features_with_9_values].apply(lambda x: (x == 9).sum()).sort_values(ascending=False)
```

```python
value_9_summary_df = value_9_counts.reset_index()
value_9_summary_df.columns = ["Feature", "Count"]
```
